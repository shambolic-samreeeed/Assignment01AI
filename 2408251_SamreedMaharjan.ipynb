*   Task - 3 - Outlier Detection:
  1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
  2. Define outliers using the 1.5  IQR rule.
  3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
in a different color.
  4. Discuss the characteristics of these outliers and their potential impact on regional averages.




```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Identifying the outlier country
#Calculating the interquartile range score and gdp using 1st and 3rd quartile
q1_score= np.percentile(southAsianDataFrame['score'],25)
q3_score =np.percentile(southAsianDataFrame['score'],75)
iqr_score = q3_score-q1_score

q1_gdp =np.percentile(southAsianDataFrame['Log GDP per capita'],25)
q3_gdp = np.percentile(southAsianDataFrame['Log GDP per capita'],75)
iqr_gdp = q3_gdp -q1_gdp

#calculating upper and lower bounds for outliers using 1.5 * IQR
lower_score_bound = q1_score -1.5 *iqr_score
upper_score_bound = q3_score+1.5 *iqr_score

lower_gdp_bound = q1_gdp -1.5 *iqr_gdp
upper_gdp_bound = q3_gdp+1.5 *iqr_gdp

outliers = southAsianDataFrame[(southAsianDataFrame['score'] <lower_score_bound) | (southAsianDataFrame['score']> upper_score_bound) |(southAsianDataFrame['Log GDP per capita']< lower_gdp_bound) | (southAsianDataFrame['Log GDP per capita']>upper_gdp_bound)]

print("Outlier Countries Based on IQR Rule:")
display(outliers[['Country name','score','Log GDP per capita']])
print()

#creating a scatter plot
plt.figure(figsize=(13,5))
plt.scatter(southAsianDataFrame['Log GDP per capita'],southAsianDataFrame['score'], label='Normal')
plt.scatter(outliers['Log GDP per capita'], outliers['score'], color='red', label='Outlier')

plt.title('Scater plot: GDP per capita and Score (red outliers)')
plt.ylabel('Score')
plt.xlabel('Log GDP per capita')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
print()

#Regional averages
regional_average_score=southAsianDataFrame['score'].mean()
regional_agerage_gdp=southAsianDataFrame['Log GDP per capita'].mean()

no_outlier_dataset = southAsianDataFrame.drop(outliers.index)
no_outlier_regional_average_score=no_outlier_dataset['score'].mean()
no_outlier_regional_agerage_gdp=no_outlier_dataset['Log GDP per capita'].mean()


print('Characterstics of the outlier country: ')
display(outliers[['Country name','score','Log GDP per capita']])
print("\nRegional Average before removing the outlier countries.")
print(f'Average Score: {regional_average_score:.2f}')
print(f'Average GDP: {regional_agerage_gdp}')
print()
print("Regional Average after removing the outlier countries.")
print(f'Average Score: {no_outlier_regional_average_score:.2f}')
print(f'Average GDP: {no_outlier_regional_agerage_gdp}')

print('\nIt can be seen that Afghanistan is the outlier country based on score and GDP,\n Removing it causes the regional average to increase. Increasing both Score and Log GDP per capita average.')
```

*   Task - 4 - Exploring Trends Across Metrics:
    1. Choose two metrics (e.g., Freedom to Make Life Choices and Generosity) and calculate their correlation
{pearson correlation} with the Score for South Asian countries
    2. Create scatter plots with trendlines for these metrics against the Score.
    3. Identify and discuss the strongest and weakest relationships between these metrics and the Score for
South Asian countries.





```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

southAsianDataFrame= pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Assessment/southAsianCountries.csv')

#Calculating pearson correalion coefficients
freedom_correlation=southAsianDataFrame['Freedom to make life choices'].corr(southAsianDataFrame['score'])
generosity_correlation = southAsianDataFrame['Generosity'].corr(southAsianDataFrame['score'])

print('Pearson Correlation Coefficients')
print(f'Freedom to make life choice and score: {freedom_correlation:.2f}')
print(f'Generosity and score: {generosity_correlation:.2f}')

#Scatter plot
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

sns.regplot(x='Freedom to make life choices',y='score', data=southAsianDataFrame,ax=axes[0],color='blue')
axes[0].set_title('Freedom to make life choices and score')
axes[0].set_xlabel('Freedom to make life choices')
axes[0].set_ylabel('Score')

sns.regplot(x='Generosity', y='score', data=southAsianDataFrame, ax=axes[1],color='green')
axes[1].set_title('Generosity and Score')
axes[1].set_xlabel('Generosity')
axes[1].set_ylabel('Score')

plt.tight_layout()
plt.show()

#analysing the relationship
if abs(freedom_correlation) > abs(generosity_correlation):
    strongest_relationship = "Freedom to make life choice"
    weakest_relationship = "Generosity"
else:
    strongest_relationship = "Generosity"
    weakest_relationship = "Freedom to make life choices"

print("\nAnalysis of the relationship:")
print(f"The strongest relationship with score is with {strongest_relationship}")
print(f"The weakest relationship with score is with {weakest_relationship}")
```

6. Outlier Detection:


*   Identify outlier countries in both regions based on Score and GDP per Capita.
*   Plot these outliers and discuss their implications.



```python
#function to calculate interquartilerange (iqr)
def getOutliers(data,column):
  q1, q3= data[column].quantile([0.25,0.75])
  iqr=q3-q1
  lowerBound=q1-1.5*iqr
  upperBound=q3+1.5*iqr
  return data[(data[column]<lowerBound)|(data[column]>upperBound)]

#using the function and caluclating outliers

outlierSouthAsianScore=getOutliers(southAsianDataFrame,'score')
outlierMiddleEastScore=getOutliers(middleEastDataFrame,'score')
outlierSouthAsiaGDP = getOutliers(southAsianDataFrame,'Log GDP per capita')
outlierMiddleEastGDP = getOutliers(middleEastDataFrame,'Log GDP per capita')

#combining the outliners
outlierSouthAsia = pd.concat([outlierSouthAsianScore,outlierSouthAsiaGDP]).drop_duplicates()
outlierMiddleEast = pd.concat([outlierMiddleEastScore, outlierMiddleEastGDP]).drop_duplicates()

plt.scatter(southAsianDataFrame['Log GDP per capita'],southAsianDataFrame['score'],label='South Asian Countries')
plt.scatter(outlierSouthAsia['Log GDP per capita'],outlierSouthAsia['score'],label ='South Asian Outliers')

plt.scatter(middleEastDataFrame['Log GDP per capita'],middleEastDataFrame['score'],label='Middle Eastern Countries')
plt.scatter(outlierMiddleEast['Log GDP per capita'],outlierMiddleEast['score'],label ='Middle Eastern Outliers')
plt.title('Outlier countries on the basis of happiness score')
plt.xlabel('Log gdp per capita')
plt.ylabel('Happiness Score')
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```
